```python


# load data
df = pd.read_csv(r"C:\Users\divya\OneDrive\Desktop\projects\OIBSIP\Task1_McDonalds\data\nutrition.csv")

# quick overview
df.head(), df.shape
```

```python
df.info()
df.describe().T
```

```python
missing = df.isnull().sum().sort_values(ascending=False)
missing
```

```python
# drop exact duplicate rows
df = df.drop_duplicates()

# fill missing nutrient values with median
for col in df.columns:
    if df[col].dtype != 'object':
        df[col] = df[col].fillna(df[col].median())

# verify all missing are handled
df.isnull().sum()
```

```python
plt.figure(figsize=(10,6))
sns.histplot(df['Calories'], bins=30, kde=True, color='crimson')
plt.title("Calories Distribution in McDonald's Menu")
plt.xlabel("Calories")
plt.ylabel("Count")
plt.show()
```

```python
plt.figure(figsize=(12,8))
sns.boxplot(x='Category', y='Calories', data=df)
plt.xticks(rotation=45)
plt.title("Calories by Menu Category")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

numeric_cols = [
    'Calories','Total Fat','Saturated Fat','Trans Fat',
    'Cholesterol','Sodium','Carbohydrates','Dietary Fiber',
    'Sugars','Protein'
]

plt.figure(figsize=(12,6))
sns.heatmap(df[numeric_cols].corr(), annot=True, linewidths=0.5)
plt.title("Nutrient Correlation Heatmap")
plt.show()
plt.savefig("images/heatmap.png")
```

```python
sns.pairplot(df[numeric_cols])
plt.show()
```

```python
plt.figure(figsize=(8,4))
sns.histplot(df['Calories'], kde=True)
plt.title("Distribution of Calories")
plt.show()
```

```python
health_score = (
    df['Calories']*0.5 +
    df['Total Fat']*0.2 +
    df['Sodium']*0.3
)

df['Health Score'] = health_score

healthiest = df.sort_values('Health Score').head(10)
healthiest
```

```python
unhealthiest = df.sort_values('Health Score', ascending=False).head(10)
unhealthiest
```

```python
healthiest.to_csv("healthiest_items.csv", index=False)
unhealthiest.to_csv("unhealthiest_items.csv", index=False)

df[numeric_cols].describe().to_csv("nutrient_stats.csv")
```
